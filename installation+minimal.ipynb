# Minimal running example

```python
import elfen
import datasets
import polars as pl
```

```python
ds = datasets.load_dataset("stanfordnlp/sst2")
```

```python
ds = pl.from_pandas(ds["test"].to_pandas())
```

```python
ds.head()
```

Test initialization

```python
extractor = elfen.Extractor(
    data=ds,
    text_column="sentence",
    language="en",
    model="en_core_web_sm",
)
```

Test extraction of features that do not require a lexicon

```python
extractor.extract_feature_group("surface")
```

```python
"n_tokens" in extractor.data.columns
```

Test extraction of features that require a lexicon that is automatically downloaded

```python
extractor.extract("n_high_concreteness")
```

```python
extractor.data["n_high_concreteness"].describe()
```

Test extraction of features that require a lexicon that has to be manually added

```python
extractor.extract("n_high_intensity")
```

```python
extractor.data["n_high_intensity_joy"].describe()
```
